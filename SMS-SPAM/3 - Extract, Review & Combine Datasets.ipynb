```python
SPAM_CLASSIFIER_DIR = DATASETS_DIR / 'Spam-Classifier'

SMS_SPAM_DIR = SPAM_CLASSIFIER_DIR / 'Sms-Spam'
SMS_SPAM_DIR.mkdir(exist_ok = True, parents = True)

YOUTUBE_SPAM_DIR = SPAM_CLASSIFIER_DIR / 'Youtube-Spam'
YOUTUBE_SPAM_DIR.mkdir(exist_ok = True, parents = True)

EXPORT_DIR = DATASETS_DIR / 'Exports'
EXPORT_DIR.mkdir(exist_ok = True, parents = True)
SPAM_DATASETS_DIR = EXPORT_DIR / 'Spam_Dataset.csv'
```

```python
import pandas as pd
```

```python
#All files contained in SMS_SPAM_DIR: 
sms_spam_input_path = SMS_SPAM_DIR / 'SMSSpamCollection'

sms_df = pd.read_csv(sms_spam_input_path, sep='\t', header=None) #Setting header None in order to include all datapoints
sms_df.columns  = ['LABEL', 'CONTENT']
sms_df['SOURCE'] = 'sms-spam'
sms_df['CLASS'] = sms_df['LABEL'].apply(lambda x: '1' if str(x) == 'spam' else '0')
sms_df = sms_df[["CONTENT", "CLASS",'LABEL', "SOURCE"]]
sms_df
```

```python
#All files contained in YOUTUBE_SPAM_DIR: 
youtube_spam_df = []

#Looping through YOUTUBE_SPAM_DIR containing a number of .csv files - furthermore, appending these to a list. 
for path in YOUTUBE_SPAM_DIR.glob('*.csv'): 

    df_raw = pd.read_csv(path)
    
    df_raw_copy = df_raw.copy()
    df_raw_copy['SOURCE'] = 'Spam-youtube'
    df_raw_copy = df_raw_copy.drop(columns=['COMMENT_ID', 'AUTHOR', 'DATE']) #Droppped irrelevant data from database
    df_raw_copy['LABEL'] = df_raw_copy['CLASS'].apply(lambda x: 'spam' if str(x) == '1' else 'ham')
    df_raw_copy = df_raw_copy[["CONTENT", "CLASS",'LABEL', "SOURCE"]]
    youtube_spam_df.append(df_raw_copy)

youtube_df =  pd.concat(youtube_spam_df, ignore_index=True) #concat, merging a list of pd.read_csv into one
youtube_df
```

```python
df_final =  pd.concat([sms_df,youtube_df])
df_final
```

```python
df_final.to_csv(SPAM_DATASETS_DIR, index=False) #Pushing Combined Dataset to .csv at SPAM_DATASETS_DIR for future work.
```
